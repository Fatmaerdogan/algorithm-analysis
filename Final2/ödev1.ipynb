```python
import random

words = ['q','w','e','r','t','y','u','o','p','a','s','d','f','g','h','j','k','l','i','z','x','c','v','b','n','m']
matrix = []
def matrixCreating(s1,s2):
    newMatrix = []
    for i in range(s1):
        newMatrix.append([])
        for j in range(s2):
            loc = random.randint(0,len(words)-1)
            newMatrix[i].append(words[loc])
    global matrix
    matrix = newMatrix
    print_matrix(matrix)
                        
def print_matrix(m):
    for i in m:
        print(i)
        
```

```python
def myReverse(v):
    v2=[]
    for i in v[::-1]:
        v2.append(i)
    return v2

def palindrom_list():
    global matrix
    if(len(matrix[0]) < 10):
        return []
    elif(len(matrix[0]) == 10):
        list = []
        for i in range(len(matrix)):          
            if(matrix[i] == myReverse(matrix[i])):
                if(matrix[i] not in list):
                    list.append(matrix[i])
        return list
    else:
        list = []
        for i in range(len(matrix)):
            a = matrix[i]
            count = len(a)-9
            count = count*(count+1)/2
            mim_len,konum = 10,0
            for j in range(int(count)):
                v = matrix[i][konum:mim_len]

                if(v == myReverse(v)):
                    if(v not in list):
                        list.append(v)
                    mim_len+=1
                else:
                    mim_len+=1
                
                if(konum+mim_len== len(matrix[0])+1):
                    mim_len= 10
                    konum+=1
        return list
```
